```python
import h5py
import glob 
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def removeDuplicates(thumbnail):
    
    seen_labels = []
    
    boxes = thumbnail["boxes"]
    scores= thumbnail["scores"]
    labels= thumbnail["labels"]
    
    new_boxes = []
    new_scores= []
    new_labels= []
    
    for idx,label in enumerate(labels):
        
        if label not in seen_labels:
            seen_labels.append(label)
            #np.append(new_boxes,boxes[idx],axis=0)
            #np.append(new_scores,scores[idx],axis=0)
            #np.append(new_labels,[label],axis=0)
            new_boxes.append(boxes[idx])
            new_scores.append(scores[idx])
            new_labels.append(label)
    
    new_boxes = np.array(new_boxes)
    new_scores= np.array(new_scores)
    new_labels= np.array(new_labels)
    
    return {"boxes":new_boxes,"scores":new_scores,"labels":new_labels}  
```

```python
COCO_INSTANCE_CATEGORY_NAMES = [
    '_background_', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'N/A', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'N/A', 'backpack', 'umbrella', 'N/A', 'N/A',
    'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball',
    'kite', 'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
    'bottle', 'N/A', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl',
    'banana', 'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
    'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'N/A', 'dining table',
    'N/A', 'N/A', 'toilet', 'N/A', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
    'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'N/A', 'book',
    'clock', 'vase', 'scissors', 'teddy bear', 'hair drier', 'toothbrush'
]

dict_keys = ['CA','DE','FR','GB','IN','JP','KR','MX','RU','US']
dset = {}
reduced_dset = {}

for f in glob.glob('data/*.{}'.format('h5')):
    with h5py.File(f, 'r') as h5_file: 
        
        class_name = os.path.splitext(os.path.basename(f))[0]
        all_thumbnails = []
        reduced_thumbnails = []
        
        for key in h5_file.keys():
            
            boxes = h5_file[key]['boxes'][:]
            scores = h5_file[key]['scores'][:]
            labels = [COCO_INSTANCE_CATEGORY_NAMES[i] for i in h5_file[key]['labels'][:]]
            
            one_thumbnail = {"boxes":boxes,"scores":scores,"labels":labels}
            
            all_thumbnails.append(one_thumbnail)
            reduced_thumbnails.append(removeDuplicates(one_thumbnail))
            
            dset[class_name] = all_thumbnails
            reduced_dset[class_name] = reduced_thumbnails
```

## Likes Analysis

```python
import re
import glob
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trendings = {}
for file in glob.glob("youtube-trending/*1.{}".format('csv')):
    country_code = re.search('youtube-trending\\\(.*)videos1.csv', file).group(1)
    print(country_code)
    
    temp = pd.read_csv(file,encoding="iso-8859-1")
    trendings[country_code] = temp.drop(columns = ["Unnamed: 0","comments_disabled","ratings_disabled","video_error_or_removed"],axis=1)
    
```

### Breakdown of categories per country

```python
genre_dict = {}
breakdowns = {}
```

```python
f = open("categoryIDtext.txt", "r")
for idx,line in enumerate(f):
    parsed = line.split(' - ')
    t1 = int(parsed[0])
    t2 = parsed[1]
    t2 = t2.replace('\n','')
    genre_dict[t1] = t2
```

```python
idx = 1
matplotlib.rcParams['figure.figsize'] = [15, 15]

fig = plt.figure()
for country,df in trendings.items():
    row_count = df.shape[0]
    breakdown = df.category_id.value_counts()/row_count
    breakdown = breakdown.rename(genre_dict)
    breakdowns[country] = breakdown
    plt.subplot(5,2,idx)
    plot = breakdown.plot.pie()
    idx = idx + 1
```

```python
averages = pd.DataFrame(index=["country"],columns=["views",'likes','dislikes','comment_count','like_per_view','dislike_per_view','comment_per_view'])
cat_averages = {} 

for country,dframe in trendings.items():
    avg_df = dframe.copy()
    avg_df = avg_df.drop(columns=['trending_date','channel_title','publish_time','tags','thumbnail_link','description','title'])
    avg_df = avg_df.groupby('category_id').mean()
        
    avg_df['like_per_view'] = avg_df['likes']/avg_df['views']
    avg_df['dislike_per_view'] = avg_df['dislikes']/avg_df['views']
    avg_df['comment_per_view'] = avg_df['comment_count']/avg_df['views']
    
    for idx in avg_df.index:
        row = avg_df[avg_df.index == idx]
        row = row.rename(index={idx:country})
        if idx in cat_averages.keys():
            cat_averages[idx] = cat_averages[idx].append(row)
        else:
            cat_averages[idx] = row
```

```python
cat_labels = {'sport':['frisbee','skis','snowboard','sports ball','kite',
                       'baseball bat','baseball glove','skateboard','surfboard','tennis racket'],
              'transportation & traffic':['bicycle','car','motorcycle','airplane','bus','train',
                                          'truck','boat','traffic light','fire hydrant','stop sign','parking meter','bench'],
              'food & drinks':['banana','apple','sandwich','orange','broccoli','carrot','hot dog','pizza','donut','cake'],
              'animal':['bird','cat','dog','horse','sheep','cow','elephant','bear','zebra','giraffe'],
              'kitchenware':['dining table','microwave','oven','sink','knife','refrigerator','spoon',
                             'bottle','wine glass','cup','bowl','fork','toaster'],
              'household':['bed','tv','laptop','remote','toilet','potted plant','umbrella','clock','vase','scissors',
                           'teddy bear','hair drier','toothbrush','chair','couch'],
              'office & belongings':['book','laptop','keyboard','mouse','tie','cell phone','suitcase','handbag','backpack'],
              'person':['person']
             }
reverseLookup = {}
              
for k,v in cat_labels.items():
    
    for elem in v:
        reverseLookup[elem] = k
```

```python
dict_keys = ['CA','DE','FR','GB','IN','JP','KR','MX','RU','US']
categories = cat_labels.keys()
cols = ['numVideos','views','likes','comments'] + list(genre_dict.values())
stats = pd.DataFrame(0,index=categories,columns=cols)
statsCountries = {}
```

```python
for f in glob.glob('stats\*.{}'.format('csv')):
    country = re.search('stats\\\(.*)stats.csv',f).group(1)
    df = pd.read_csv(f,index_col=0)
    (df != 0).any(axis=0)
    df = df.loc[:, (df != 0).any(axis=0)]
    df.columns = ['Comedy' if x=='Comedy.1' else x for x in df.columns]
    statsCountries[country] = df
```

```python
temp = pd.DataFrame(0,index=statsCountries.keys(),columns=cat_labels.keys())
ViewsAvgs = temp.copy()
LikesAvgs = temp.copy()
CommentsAvgs = temp.copy()
```

```python
for country,df in statsCountries.items():
    ViewsAvgs.loc[country,:] = df.loc[:,"views"]/df.loc[:,'numVideos']
    LikesAvgs.loc[country,:] = df.loc[:,"likes"]/df.loc[:,'numVideos']
    CommentsAvgs.loc[country,:] = df.loc[:,"comments"]/df.loc[:,'numVideos']
```
